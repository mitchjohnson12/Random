```python
import pandas as pd

path = '/Users/mitchjohnson/Documents/NPPES/NPPES_Data_Dissemination_January_2024/npidata_pfile_20050523-20240107.csv'
cols = ['Entity Type Code', 'NPI', 'Provider Organization Name (Legal Business Name)', 'Provider Enumeration Date', 'Last Update Date', 'NPI Deactivation Date', 'NPI Reactivation Date', 'Healthcare Provider Taxonomy Code_1', 'Healthcare Provider Taxonomy Code_2']
df_nppes = pd.read_csv(path, usecols=cols)
df_nppes.rename(columns={'Entity Type Code':'code', 'Provider Organization Name (Legal Business Name)':'OrgName'}, inplace=True)
df_nppes
```

```python
df_nppes.shape
#df[df.ein != '<UNAVAIL>']
```

```python
path2 = '/Users/mitchjohnson/Documents/NPPES/inputs/Practice NPIs from Richard.csv'
df_richard = pd.read_csv(path2)
df_richard
```

```python
df_richard_nppes = pd.merge(df_nppes, df_richard, on='NPI', how='inner')
df_richard_nppes.shape
```

```python
df_richard_nppes.groupby('code').size()
```

```python
richard_pathname = '/Users/mitchjohnson/Documents/NPPES/outputs/richard_type_1s.csv'
df_richard_nppes[df_richard_nppes.code == 1.0].NPI.to_csv(richard_pathname, index=False)
```
